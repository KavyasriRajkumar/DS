```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Fraud Detection Dataset.csv")

# ---- SIMPLE PREPROCESSING ----
# Fill missing values with 0
df = df.fillna(0)

# Convert all text columns into numeric codes
df = df.apply(lambda x: x.astype('category').cat.codes)

print("Processed Data:")
print(df.head())

# ---- SIMPLE VISUALIZATION ----
# 1. Distribution of Transaction Amount
plt.figure(figsize=(6,4))
sns.histplot(df["Transaction_Amount"], kde=True)
plt.title("Transaction Amount Distribution")
plt.show()

# 2. Fraudulent vs Non-Fraudulent Count
plt.figure(figsize=(5,4))
sns.countplot(x=df["Fraudulent"])
plt.title("Fraudulent vs Non-Fraudulent")
plt.show()

# 3. Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=False, cmap="viridis")
plt.title("Correlation Heatmap")
plt.show()
```
